# Space Lattices in Semiconductor Physics

## 1. Introduction

A crystal lattice is a periodic arrangement of atoms or molecules in a crystalline material. Understanding space lattices is fundamental to semiconductor physics as the arrangement of atoms directly influences the electronic, optical, and thermal properties of semiconductor materials.

In this notebook, we'll explore:
- Basic concepts of crystal lattices
- Bravais lattices and crystal systems
- Miller indices
- Common semiconductor crystal structures
- Visualization and analysis techniques


## 2. Theoretical Background

### 2.1 Definitions

A **space lattice** (or Bravais lattice) is an infinite array of discrete points with an arrangement and orientation that appears exactly the same from any of the points in the array.

Mathematically, we can describe a lattice by:

$$\vec{r} = n_1\vec{a}_1 + n_2\vec{a}_2 + n_3\vec{a}_3$$

Where:
- $\vec{r}$ is the position vector of any lattice point
- $\vec{a}_1, \vec{a}_2, \vec{a}_3$ are the primitive vectors
- $n_1, n_2, n_3$ are integers

### 2.2 Crystal Systems and Bravais Lattices

There are 7 crystal systems and 14 Bravais lattices in three dimensions:

- Cubic (Simple, Body-Centered, Face-Centered)
- Tetragonal (Simple, Body-Centered)
- Orthorhombic (Simple, Body-Centered, Face-Centered, Base-Centered)
- Monoclinic (Simple, Base-Centered)
- Triclinic
- Hexagonal
- Rhombohedral

## 3. Visualization of Common Lattice Structures
Let's create visualizations for the most important lattice structures in semiconductor physics.

### 3.1 Simple Cubic Lattice

````python
def plot_simple_cubic(size=2):
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='3d')
    
    # Create points
    for x in range(size):
        for y in range(size):
            for z in range(size):
                ax.scatter(x, y, z, color='blue', s=5000)
    
    # Set labels and limits
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_xlim(-0.5, size-0.5)
    ax.set_ylim(-0.5, size-0.5)
    ax.set_zlim(-0.5, size-0.5)
    ax.set_title('Simple Cubic Lattice')
    
    # Add bonds between adjacent atoms
    for x in range(size):
        for y in range(size):
            for z in range(size):
                if x < size-1:
                    ax.plot([x, x+1], [y, y], [z, z], 'k-', alpha=0.5)
                if y < size-1:
                    ax.plot([x, x], [y, y+1], [z, z], 'k-', alpha=0.5)
                if z < size-1:
                    ax.plot([x, x], [y, y], [z, z+1], 'k-', alpha=0.5)
    
    plt.tight_layout()
    return fig
````

````python

# Display the simple cubic lattice
plot_simple_cubic(3)
````

## 4. Miller Indices

Miller indices are a notation system used to specify planes and directions in crystal lattices.

### 4.1 Definition and Calculation

To determine the Miller indices (h, k, l) of a plane:
1. Find the intercepts of the plane with the axes, in terms of the lattice constants
2. Take the reciprocals of these intercepts
3. Reduce to the smallest set of integers by multiplying by a common factor

```python
def plot_miller_indices(h, k, l, size=2):
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='3d')
    
    # Plot simple cubic lattice
    for x in range(size+1):
        for y in range(size+1):
            for z in range(size+1):
                ax.scatter(x, y, z, color='blue', s=50)
    
    # Calculate the plane
    max_val = size
    if h != 0 and k != 0 and l != 0:
        # Generate a grid of points
        xx, yy = np.meshgrid(np.linspace(0, max_val, 10), np.linspace(0, max_val, 10))
        
        # Calculate z from the plane equation: h*x + k*y + l*z = n
        # We set n = h*k*l (one possible value)
        n = max_val * h * k * l / max([h, k, l])
        zz = (n - h*xx - k*yy) / l if l != 0 else np.zeros_like(xx)
        
        # Plot the plane
        ax.plot_surface(xx, yy, zz, alpha=0.3, color='red')
        
    elif l == 0:  # Plane parallel to z-axis
        xx, zz = np.meshgrid(np.linspace(0, max_val, 10), np.linspace(0, max_val, 10))
        n = max_val * h * k / max([h, k, 1])
        yy = (n - h*xx) / k if k != 0 else np.zeros_like(xx)
        ax.plot_surface(xx, yy, zz, alpha=0.3, color='red')
        
    elif k == 0:  # Plane parallel to y-axis
        xx, zz = np.meshgrid(np.linspace(0, max_val, 10), np.linspace(0, max_val, 10))
        n = max_val * h * l / max([h, l, 1])
        yy = np.zeros_like(xx)
        ax.plot_surface(xx, yy, zz, alpha=0.3, color='red')
        
    elif h == 0:  # Plane parallel to x-axis
        yy, zz = np.meshgrid(np.linspace(0, max_val, 10), np.linspace(0, max_val, 10))
        n = max_val * k * l / max([k, l, 1])
        xx = np.zeros_like(yy)
        ax.plot_surface(xx, yy, zz, alpha=0.3, color='red')
    
    # Set labels and limits
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_xlim(0, max_val)
    ax.set_ylim(0, max_val)
    ax.set_zlim(0, max_val)
    ax.set_title(f'Miller Indices: ({h},{k},{l})')
    
    plt.tight_layout()
    return fig

# Example: (1, 1, 1) plane in cubic lattice
plot_miller_indices(1, 1, 1)
```

import ipywidgets as widgets

def plot_direction_property(direction='[100]'):
    # Data
    directions = ['[100]', '[110]', '[111]']

    # Plot

    ax.set_ylabel('Electron Mobility (cm/Vs)')
    ax.set_title(f'Electron Mobility in Silicon for Different Crystal Directions')

## 7. X-ray Diffraction and Reciprocal Lattice

X-ray diffraction is a key technique for crystal structure determination, governed by Bragg's law:

$$n\lambda = 2d\sin\theta$$

Where:
- $n$ is an integer
- $\lambda$ is the wavelength of the X-rays
- $d$ is the spacing between planes
- $\theta$ is the angle of incidence
