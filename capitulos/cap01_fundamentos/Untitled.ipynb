```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Markdown, HTML
import ipywidgets as widgets

# Para estructuras cristalinas y visualizaciÃ³n
from ase import Atom, Atoms
from ase.visualize import view
from ase.lattice.cubic import Diamond, FaceCenteredCubic
# from ase.lattice.hexagonal import Wurtzite
import nglview as nv

# Para simulaciones y cÃ¡lculos
from pymatgen.core.structure import Structure
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python

# Para anÃ¡lisis de bandas (opcional, requiere instalaciÃ³n adicional)
# import pymatgen.electronic_structure as es

# FunciÃ³n para crear estructuras semiconductoras comunes
def create_semiconductor_structure(material='Si', size=(2, 2, 2)):
    """
    Crea estructuras cristalinas de materiales semiconductores comunes.
    
    ParÃ¡metros:
    material (str): Tipo de material semiconductor ('Si', 'Ge', 'GaAs', 'GaN', etc.)
    size (tuple): TamaÃ±o de la supercelda (nx, ny, nz)
    
    Retorna:
    ase.Atoms: Estructura cristalina del material
    """
    # Definir parÃ¡metros de red para semiconductores comunes (Ã)
    params = {
        'Si': {'type': 'diamond', 'a': 5.431, 'elements': ['Si']},
        'Ge': {'type': 'diamond', 'a': 5.658, 'elements': ['Ge']},
        'GaAs': {'type': 'zincblende', 'a': 5.653, 'elements': ['Ga', 'As']},
        'InP': {'type': 'zincblende', 'a': 5.869, 'elements': ['In', 'P']},
        'GaN': {'type': 'wurtzite', 'a': 3.189, 'c': 5.185, 'elements': ['Ga', 'N']},
        'ZnO': {'type': 'wurtzite', 'a': 3.250, 'c': 5.207, 'elements': ['Zn', 'O']}
    }
    
    # Verificar si el material estÃ¡ soportado
    if material not in params:
        raise ValueError(f"Material '{material}' no soportado. Opciones: {list(params.keys())}")
    
    material_params = params[material]
    
    # Crear la estructura segÃºn el tipo
    if material_params['type'] == 'diamond':
        # Estructura diamante (Si, Ge)
        structure = Diamond(
            symbol=material_params['elements'][0],
            latticeconstant=material_params['a'],
            size=size
        )
    
    elif material_params['type'] == 'zincblende':
        # Estructura zinc blenda (GaAs, InP, etc.)
        # ASE no tiene una clase especÃ­fica para zinc blenda, usamos FCC con base
        a = material_params['a']
        element_a, element_b = material_params['elements']
        
        # Crear estructura FCC para el primer elemento
        structure = FaceCenteredCubic(
            symbol=element_a,
            latticeconstant=a,
            size=size
        )
        
        # AÃ±adir el segundo elemento en las posiciones desplazadas (0.25, 0.25, 0.25)
        positions = structure.get_scaled_positions()
        cell = structure.get_cell()
        
        # Crear desplazamiento para el segundo elemento
        displacement = np.array([0.25, 0.25, 0.25])
        
        # AÃ±adir los Ã¡tomos del segundo elemento
        for i in range(len(positions)):
            pos = (positions[i] + displacement) % 1.0  # Aplicar periodicidad
            structure.append(Atom(element_b, np.dot(pos, cell)))
    
    elif material_params['type'] == 'wurtzite':
        # Estructura wurtzita (GaN, ZnO, etc.)
        a = material_params['a']
        c = material_params['c']
        element_a, element_b = material_params['elements']
        
        structure = Wurtzite(
            symbol=element_a,
            latticeconstant={'a': a, 'c': c},
            size=size
        )
        
        # AÃ±adir el segundo elemento en las posiciones desplazadas
        positions = structure.get_scaled_positions()
        cell = structure.get_cell()
        
        # La mitad de los Ã¡tomos deben ser del segundo elemento
        n_atoms = len(positions)
        structure.symbols[n_atoms//2:] = element_b
    
    return structure

# FunciÃ³n para visualizar estructura con informaciÃ³n
def visualize_semiconductor(material='Si', size=(2, 2, 2), show_bonds=True):
    """
    Visualiza una estructura cristalina semiconductora con informaciÃ³n relevante.
    
    ParÃ¡metros:
    material (str): Tipo de material semiconductor
    size (tuple): TamaÃ±o de la supercelda
    show_bonds (bool): Mostrar enlaces entre Ã¡tomos
    
    Retorna:
    nglview.NGLWidget: Visualizador interactivo de la estructura
    """
    # Crear la estructura
    structure = create_semiconductor_structure(material, size)
    
    # Mostrar informaciÃ³n del material
    display(Markdown(f"## Estructura cristalina de {material}"))
    
    # Obtener informaciÃ³n de la estructura
    num_atoms = len(structure)
    volume = structure.get_volume()
    cell = structure.get_cell()
    a = np.linalg.norm(cell[0])
    
    # Determinar el tipo de estructura
    structure_type = "Desconocida"
    if material in ['Si', 'Ge']:
        structure_type = "Diamante (cÃºbica)"
    elif material in ['GaAs', 'InP']:
        structure_type = "Zinc blenda (cÃºbica)"
    elif material in ['GaN', 'ZnO']:
        structure_type = "Wurtzita (hexagonal)"
    
    # Mostrar informaciÃ³n bÃ¡sica
    info_html = f"""
    <div style="background-color: #f0f0f0; padding: 10px; border-radius: 5px; margin-bottom: 20px;">
        <h3>InformaciÃ³n del material</h3>
        <ul>
            <li><b>Material:</b> {material}</li>
            <li><b>Estructura:</b> {structure_type}</li>
            <li><b>ParÃ¡metro de red:</b> {a:.3f} Ã</li>
            <li><b>NÃºmero de Ã¡tomos:</b> {num_atoms}</li>
            <li><b>Volumen:</b> {volume:.2f} ÃÂ³</li>
        </ul>
    </div>
    """
    
    display(HTML(info_html))
    
    # Crear visualizaciÃ³n
    view = nv.show_ase(structure)
    
    # Configurar visualizaciÃ³n
    if show_bonds:
        view.add_ball_and_stick()
    
    view.add_unitcell()
    view.center()
    
    # AÃ±adir informaciÃ³n sobre aplicaciones del material
    applications = {
        'Si': "Microprocesadores, cÃ©lulas solares, dispositivos de potencia",
        'Ge': "Transistores de alta velocidad, detectores infrarrojos",
        'GaAs': "Dispositivos optoelectrÃ³nicos, LEDs, lÃ¡seres, cÃ©lulas solares de alta eficiencia",
        'InP': "FotÃ³nica, dispositivos de alta frecuencia, comunicaciones Ã³pticas",
        'GaN': "LEDs azules, dispositivos de potencia, alta temperatura",
        'ZnO': "Sensores, dispositivos UV, piezoelÃ©ctricos"
    }
    
    if material in applications:
        display(Markdown(f"**Aplicaciones principales de {material}**: {applications[material]}"))
    
    # Mostrar informaciÃ³n sobre propiedades electrÃ³nicas
    bandgaps = {
        'Si': 1.12,
        'Ge': 0.67,
        'GaAs': 1.42,
        'InP': 1.35,
        'GaN': 3.4,
        'ZnO': 3.37
    }
    
    if material in bandgaps:
        display(Markdown(f"**Ancho de banda prohibida (bandgap)**: {bandgaps[material]} eV"))
    
    return view

# Widget interactivo para explorar diferentes materiales semiconductores
def interactive_semiconductor_explorer():
    """
    Crea un widget interactivo para explorar diferentes materiales semiconductores.
    """
    # Widgets para controlar los parÃ¡metros
    material_dropdown = widgets.Dropdown(
        options=['Si', 'Ge', 'GaAs', 'InP', 'GaN', 'ZnO'],
        value='Si',
        description='Material:',
    )
    
    size_slider = widgets.IntSlider(
        value=2,
        min=1,
        max=4,
        step=1,
        description='TamaÃ±o:',
        disabled=False,
        continuous_update=False
    )
    
    bonds_checkbox = widgets.Checkbox(
        value=True,
        description='Mostrar enlaces',
        disabled=False
    )
    
    # Widget para mostrar la visualizaciÃ³n
    output = widgets.Output()
    
    # FunciÃ³n para actualizar la visualizaciÃ³n
    def update_view(material, size, show_bonds):
        output.clear_output()
        with output:
            visualize_semiconductor(
                material=material, 
                size=(size, size, size),
                show_bonds=show_bonds
            )
    
    # Conectar widgets con la funciÃ³n de actualizaciÃ³n
    ui = widgets.interactive(
        update_view, 
        material=material_dropdown,
        size=size_slider,
        show_bonds=bonds_checkbox
    )
    
    # Organizar la interfaz de usuario
    controls = widgets.VBox([
        widgets.HBox([material_dropdown, size_slider, bonds_checkbox])
    ])
    
    # Mostrar widgets
    display(controls)
    display(output)
    
    # Iniciar la visualizaciÃ³n con los valores predeterminados
    update_view(material_dropdown.value, size_slider.value, bonds_checkbox.value)

# FunciÃ³n para simular y visualizar planos cristalinos importantes en semiconductores
def visualize_crystal_planes(material='Si'):
    """
    Visualiza planos cristalogrÃ¡ficos importantes en semiconductores.
    
    ParÃ¡metros:
    material (str): Material semiconductor a visualizar
    """
    # Crear estructura base
    structure = create_semiconductor_structure(material, size=(2, 2, 2))
    
    # Definir planos importantes para cada material
    important_planes = {
        'Si': [(1, 0, 0), (1, 1, 0), (1, 1, 1)],
        'Ge': [(1, 0, 0), (1, 1, 0), (1, 1, 1)],
        'GaAs': [(1, 0, 0), (1, 1, 0), (1, 1, 1)],
        'GaN': [(0, 0, 0, 1), (1, 1, 0, 0), (1, 0, -1, 0)]  # NotaciÃ³n para wurtzita
    }
    
    planes = important_planes.get(material, [(1, 0, 0), (1, 1, 0), (1, 1, 1)])
    
    # Mostrar informaciÃ³n sobre los planos
    display(Markdown(f"## Planos cristalogrÃ¡ficos importantes en {material}"))
    
    # Crear visualizaciones (simuladas) para cada plano
    for plane in planes:
        plane_str = ''.join([str(i) for i in plane])
        display(Markdown(f"### Plano ({plane_str})"))
        
        # AquÃ­ normalmente habrÃ­a cÃ³digo para visualizar el plano especÃ­fico
        # En una implementaciÃ³n real se usarÃ­a ASE o PyMatGen para mostrar los planos
        
        view = nv.show_ase(structure)
        view.add_unitcell()
        
        # Nota: En una implementaciÃ³n completa, aquÃ­ se aÃ±adirÃ­a cÃ³digo para
        # mostrar el plano especÃ­fico dentro de la estructura
        
        display(view)
        
        # Mostrar informaciÃ³n sobre la importancia del plano
        if material in ['Si', 'Ge']:
            if plane == (1, 0, 0):
                display(Markdown("""
                **Importancia del plano (100)**: 
                - Usado en la fabricaciÃ³n de circuitos CMOS
                - Menor densidad de estados de interfaz
                - Menor corriente de fuga
                """))
            elif plane == (1, 1, 0):
                display(Markdown("""
                **Importancia del plano (110)**:
                - Mayor movilidad de electrones
                - Usado en dispositivos de alta velocidad
                """))
            elif plane == (1, 1, 1):
                display(Markdown("""
                **Importancia del plano (111)**:
                - Plano de exfoliaciÃ³n natural
                - Usado en crecimiento epitaxial
                - Densidad atÃ³mica mÃ¡s alta
                """))

# Ejemplo de uso para simulaciones XRD de materiales semiconductores
def simulate_xrd_pattern(material='Si', radiation='CuKa'):
    """
    Simula un patrÃ³n de difracciÃ³n de rayos X para un material semiconductor.
    
    ParÃ¡metros:
    material (str): Material semiconductor
    radiation (str): Tipo de radiaciÃ³n (CuKa, MoKa, etc.)
    """
    # Crear estructura ASE y convertir a PyMatGen
    ase_structure = create_semiconductor_structure(material)
    
    # Convertir a estructura PyMatGen
    lattice = ase_structure.get_cell()
    species = ase_structure.get_chemical_symbols()
    coords = ase_structure.get_scaled_positions()
    
    structure = Structure(
        lattice=lattice,
        species=species,
        coords=coords,
        coords_are_cartesian=False
    )
    
    # Calcular patrÃ³n XRD
    calculator = XRDCalculator(wavelength=radiation)
    pattern = calculator.get_pattern(structure, two_theta_range=(5, 90))
    
    # Visualizar patrÃ³n
    plt.figure(figsize=(10, 6))
    plt.plot(pattern.x, pattern.y)
    plt.xlabel(r'2$\theta$ (grados)')
    plt.ylabel('Intensidad (u.a.)')
    plt.title(f'PatrÃ³n de difracciÃ³n de rayos X de {material}')
    plt.grid(True, linestyle='--', alpha=0.5)
    
    # AÃ±adir etiquetas a los picos principales
    peak_indices = np.where(pattern.y > 0.1 * np.max(pattern.y))[0]
    for idx in peak_indices:
        plt.text(
            pattern.x[idx], 
            pattern.y[idx] + 0.02 * np.max(pattern.y),
            f'({pattern.hkls[idx][0][0]}{pattern.hkls[idx][0][1]}{pattern.hkls[idx][0][2]})',
            ha='center', va='bottom', fontsize=8
        )
    
    plt.tight_layout()
    plt.show()
    
    # Mostrar informaciÃ³n sobre interpretaciÃ³n del patrÃ³n
    display(Markdown(f"""
    ### InterpretaciÃ³n del patrÃ³n XRD de {material}
    
    Los picos de difracciÃ³n corresponden a las familias de planos cristalogrÃ¡ficos en la estructura.
    La posiciÃ³n de los picos (Ã¡ngulo 2Î¸) estÃ¡ relacionada con el espaciado interplanar mediante la ley de Bragg:
    
    $$n\\lambda = 2d\\sin\\theta$$
    
    donde:
    - $\\lambda$ es la longitud de onda de los rayos X
    - $d$ es la distancia interplanar
    - $\\theta$ es el Ã¡ngulo de difracciÃ³n
    
    La intensidad de los picos depende de:
    - El factor de estructura
    - La multiplicidad del plano
    - El factor de Lorentz-polarizaciÃ³n
    - El factor de temperatura
    
    Este patrÃ³n de difracciÃ³n es caracterÃ­stico de la estructura {material} y se usa para:
    - IdentificaciÃ³n de fases cristalinas
    - AnÃ¡lisis de tamaÃ±o de cristalita
    - DeterminaciÃ³n de tensiones y deformaciones
    - Control de calidad en la fabricaciÃ³n de dispositivos semiconductores
    """))

# Ejecutar el explorador interactivo
interactive_semiconductor_explorer()
```
